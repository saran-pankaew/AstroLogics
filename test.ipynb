```python
import pyboolnet

from pyboolnet.file_exchange import bnet2primes, primes2bnet
from pyboolnet.prime_implicants import find_constants, create_variables
from pyboolnet.repository import get_primes

from pyboolnet import state_transition_graphs as STGs
```

```python
import networkx as nx
import pandas as pd
```

```python
# Manual model definition
bnet = """
v1,	!v1
v2,	1
v3,	v2 & (!v1 | v3)
"""

primes = bnet2primes('./2018_zanudo_proliferation.bnet')
```

```python
update = 'asynchronous'

stg = STGs.primes2stg(primes, update)
stg.graph["rankdir"] = "LR"

#STGs.stg2image(stg, "example11_stg.pdf")
```

## Obtaining state transtion graph

```python
stg_matrix = nx.adjacency_matrix(stg)
stg_matrix = pd.DataFrame.sparse.from_spmatrix(stg_matrix)

stg_matrix.columns = list(stg.nodes)
stg_matrix.index = list(stg.nodes)
#
```
